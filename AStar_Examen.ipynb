```python
#Creacion de la clase CostoEstacionSO
#Esta clase se encarga de almacenar los datos de una estacion, como nombre y costo
#En la version original estaba definido como enum e int, para python se opto por string e int

class CostoEstacionSO(object):
    def __init__(self, estacion:str, costo:int):
        self.Estacion = estacion
        self.costo = costo

    def __str__(self):
        return f"{self.Estacion}"

testCostoEstacionSo1 = CostoEstacionSO("oceania",5)
testCostoEstacionSo2 = CostoEstacionSO("pantitlan",2)
print(testCostoEstacionSo1)
print(testCostoEstacionSo2)
```

```python
#Creacion de la clase CostoCaminoSO
#Esta clase se encarga de almacenar 2 estaciones de la clase "CostoEstacionSO" y el costo entre ellas 2

class CostoCaminoSO :
    #EstacionesSO #CostoEstacionSO[]
    #costo # int

    def __init__(self, Estaciones, costo:int):
        self.EstacionesSO : list[CostoEstacionSO] = Estaciones
        self.costo : int = costo

    def __str__(self):
        return f"({self.EstacionesSO[0].Estacion} a {self.EstacionesSO[1].Estacion}): {self.costo}"

    def GetEstaciones(self):
        return self.EstacionesSO;

testCostoCaminoSo = CostoCaminoSO([testCostoEstacionSo1, testCostoEstacionSo2],5)
print(testCostoCaminoSo)
```

```python
#Creacion de clase CaminoRecorrido
#Esta clase almacena datos que se utilizaran durante cada ciclo
#se genera un objeto de este tipo con variables actualizadas en cada ciclo

class CaminoRecorrido(object):

    def __init__(self, nodoSO:CostoEstacionSO, caminosPorRecorrer:list[CostoCaminoSO], ListaEstacionesRecorridasEnEsteCiclo:list[CostoEstacionSO], CostoCaminoRecorrido : int = 0):
        #EstacionInicialSO almacena la una estacion, la cual se revisara durante cada ciclo
        self.EstacionInicialSO = nodoSO #CostoEstacionSO

        #CostoCaminoRecorrido almacena el costo de todos los caminos recorridos hasta este momento
        self.CostoCaminoRecorrido = CostoCaminoRecorrido  #int

        #CaminosRestantesPorRecorrer almacena todos los posibles caminos que no se han recorrido de esta estacion (nodo)
        self.CaminosRestantesPorRecorrer = caminosPorRecorrer #List<CostoCaminoSO>

        #CaminosYaRecorridos almacena los ya recorridos
        self.CaminosYaRecorridos = [] #List<CostoCaminoSO>

        #Esta lista almacena todas las estaciones por las que han pasado para llegar a esta
        #sirve para poder revisar el camino completo al final
        self.ListaEstacionesRecorridasEnEsteCiclo = ListaEstacionesRecorridasEnEsteCiclo #List<CostoEstacionSO>

    #def DetenerRecorrido(self)
    #{
    #    self.PuedeSeguir = false
    #}

    def GetPuedeSeguir(self):
        return self.PuedeSeguir

    def GetEstacionActualSO(self):
        return self.EstacionInicialSO

    def SetEstacionActual(self, Estacion):
        self.EstacionInicialSO = Estacion
        self.estacionActualVisual = EstacionInicialSO.Estacion  #EstacionesConTransbordes(enum)

    def AddToCostoCaminoRecorrido(self, Costo):
        self.CostoCaminoRecorrido += Costo

    def __str__(self):
        return "Estacion actual: {}, costo del camino recorrido: {}".format(self.EstacionInicialSO, self.CostoCaminoRecorrido)
```

Para la practica, se dise√±o la clase "SistemaMetro"

```python
# Creacion de clase SistemaMetro
#Esta clase se encarga de realizar todos los calculos de A*

class SistemaMetro():

    #Datos
    #estacionesObjetivo = [] #EstacionesConTransbordes(enum)[] convertido a str[]

    #estaciones = []#List<CostoEstacionSO>
    #caminos = []#List<CostoCaminoSO>

    #Variables
    #EstacionARevisarEnElSiguienteCiclo = None#CaminoRecorrido
    #EstacionesVisitadas = []#List<CaminoRecorrido>
    #caminosPreviosYaRevisados = []#List<CostoCaminoSO>

    #Variables de ciclo
    #MaxIterations = 1#int

    #StepByStep = False#bool
    #Step = False #bool
    #SSIterationsVisual = 0 #int

    def __init__(self, estacionesObjetivo:dict[str], estaciones:CostoEstacionSO, caminos:CostoCaminoSO, EstacionARevisarEnElSiguienteCiclo:CaminoRecorrido, Iteraciones:int = 50):

        self.__PopulateClass(estacionesObjetivo, estaciones, caminos, EstacionARevisarEnElSiguienteCiclo)

        #Variables de ciclo
        self.MaxIterations = Iteraciones#int

        self.StepByStep = False#bool
        self.Step = False #bool
        self.SSIterationsVisual = 0 #int


        #RevisarDuplicados(estaciones);
        #RevisarCaminosDuplicados(caminos);

        #Aqui se encarga de revisar si el ciclo de A* llego a un resultado antes de un numero de iteraciones definido
        if(not self.StepByStep):
            for i in range(self.MaxIterations):
                self.breakVar = self.AStarCycle(i)
                if self.breakVar:
                    break

    #Esta funcion llena las variables de la clase
    def __PopulateClass(self, estacionesObjetivo:list[str], estaciones:list[CostoEstacionSO], caminos:list[CostoCaminoSO], EstacionARevisarEnElSiguienteCiclo:CaminoRecorrido):
        #Datos
        self.estacionesObjetivo = estacionesObjetivo#EstacionesConTransbordes(enum)[] convertido a str[]
        self.estaciones:list[CostoEstacionSO] = estaciones#List<CostoEstacionSO>
        self.caminos:list[CostoCaminoSO] = caminos#List<CostoCaminoSO>
        #Variables
        self.EstacionARevisarEnElSiguienteCiclo = EstacionARevisarEnElSiguienteCiclo #CaminoRecorrido
        self.EstacionesVisitadas = []#List<CaminoRecorrido>
        self.caminosPreviosYaRevisados = []#List<CostoCaminoSO>

    #Esta funcion revisa si en cada ciclo de A*, se encuentra la solucion, si si, imprime cual es, el costo
    #del camino, las iteraciones y las estaciones recorridas con colores, retorna true si encontro la solucion
    def AStarCycle(self, iterations:int)->bool:
        self.EncontroSolucion = self.AStar()
        if self.EncontroSolucion:
            log = "\033[94m found after {} iterations \033[0m \nCamino: \033[94m -".format(iterations)
            for i in range(len(self.EstacionARevisarEnElSiguienteCiclo.ListaEstacionesRecorridasEnEsteCiclo)):
            #for(int i = 0; i < self.EstacionARevisarEnElSiguienteCiclo.ListaEstacionesRecorridasEnEsteCiclo.Count; i++)
                log += "{}-".format(self.EstacionARevisarEnElSiguienteCiclo.ListaEstacionesRecorridasEnEsteCiclo[i].Estacion)
            log += "{}-".format(self.EstacionARevisarEnElSiguienteCiclo.EstacionInicialSO.Estacion)
            log += "\033[0m \nCosto del camino total sin nodo: \033[93m {}".format(self.EstacionARevisarEnElSiguienteCiclo.CostoCaminoRecorrido)
            log += "\033[0m \n\t\t\tcon nodo: \033[93m {}".format(self.EstacionARevisarEnElSiguienteCiclo.CostoCaminoRecorrido + self.NuevaEstacion.costo)
            print(log)
            return True
        return False

    # Pasos A*
    #
    # 1.- Obtener nuevos nodos hijo, agregarlos a un arreglo de todos los caminos posibles
    # 2.- Revisar de dicho arreglo, cual es el nodo con menor costo
    # 3.- Moverse al nodo con menor costo, almacenar el costo del camino hasta ese momento y el camino seguido
    # 4.- Verificar si ese nodo es el objetivo, si si terminoar, si no, repetir
    #

    #funcion definida de la implementacion de A*, se ejecuta 1 vez por ciclo
    def AStar(self) -> bool:

        print("\033[94m-------------------------\033[0m")

        #1.- Obtener nodos hijo, agregarlos a un arreglo de todos los caminos posibles
        self.BuscarCaminosEnHijosNuevos()

        #2.- Revisar de dicho arreglo, cual es el nodo con menor costo
        #PosibleOtraEstacion : CostoEstacionSO = None
        #caminoMenosCostoso: CostoCaminoSO = None
        #NuevaEstacion: CostoEstacionSO = None
        #caminoMejor: CaminoRecorrido = None
        self.PosibleOtraEstacion, self.caminoMenosCostoso, self.NuevaEstacion, self.caminoMejor = self.ObtenerCaminoDeMenorCosto()

        #debug
        #print(f"Camino menos costoso: {self.caminoMenosCostoso}\nNueva estacion: {self.NuevaEstacion}\nCamino mejor: [{self.caminoMejor}]\nEstacion actual SO: {self.EstacionARevisarEnElSiguienteCiclo.EstacionInicialSO}");

        #3.- Moverse al nodo con menor costo, almacenar el costo del camino hasta ese momento y el camino seguido
        self.MoverseAOtroNodo(self.caminoMenosCostoso, self.NuevaEstacion, self.caminoMejor);

        #4.- Verificar si ese nodo es el objetivo, si si terminoar, si no, repetir
        #for (int i = 0; i < estacionesObjetivo.Length; i++)
        for i in range(len(self.estacionesObjetivo)):
            self.estacionObjetivo = self.estacionesObjetivo[i]
            if self.estacionObjetivo == self.EstacionARevisarEnElSiguienteCiclo.EstacionInicialSO.Estacion:
                return True
        return False

    def BuscarCaminosEnHijosNuevos(self):
        self.index : int = -1
        self.containsCamino : bool = False

        for Camino in self.caminos:
            #aqui, se guardan en una variable las estaciones que estan definidas en la clase "CostoCaminoSO"
            self.estacionesQueTieneUnCamino = Camino.GetEstaciones();

            #si el camino no tiene la estacion en la que estamos, ignorar y continuar la busqueda
            if not self.EstacionARevisarEnElSiguienteCiclo.EstacionInicialSO in self.estacionesQueTieneUnCamino:
                continue

            #si el camino ya fue revisado, ignorar y continuar
            if Camino in self.caminosPreviosYaRevisados:
                continue

            #revisa si ya agregamos el camino, se ejecuta 1 vez por ciclo para agregar el camino a la
            #lista de estaciones visitadas, generando un objeto "CaminoRecorrido"
            if not self.containsCamino:
                #CopyListaEstacionesRecorridasEnEsteCiclo : list[CostoEstacionSO]
                self.CopyListaEstacionesRecorridasEnEsteCiclo = []
                for estacionRecorrida in self.EstacionARevisarEnElSiguienteCiclo.ListaEstacionesRecorridasEnEsteCiclo:

                    self.CopyListaEstacionesRecorridasEnEsteCiclo.append(estacionRecorrida)

                self.CopyListaEstacionesRecorridasEnEsteCiclo.append(self.EstacionARevisarEnElSiguienteCiclo.EstacionInicialSO)
                self.camRecorrido:CaminoRecorrido = CaminoRecorrido(self.EstacionARevisarEnElSiguienteCiclo.EstacionInicialSO, [], self.CopyListaEstacionesRecorridasEnEsteCiclo, self.EstacionARevisarEnElSiguienteCiclo.CostoCaminoRecorrido)
                self.camRecorrido.CaminosRestantesPorRecorrer.append(Camino)
                self.EstacionesVisitadas.append(self.camRecorrido)
                self.index = len(self.EstacionesVisitadas) - 1
                self.containsCamino = True
            else:
                #si el camino ya esta dentro de caminos restantes por recorrer o dentro de caminos ya recorridos, ignorar
                if Camino not in self.EstacionesVisitadas[self.index].CaminosRestantesPorRecorrer and Camino not in self.EstacionesVisitadas[self.index].CaminosYaRecorridos:
                    self.EstacionesVisitadas[self.index].CaminosRestantesPorRecorrer.append(Camino)
        #imprime los caminos restantes por recorrer en toda la lista de
        #EstacionesVisitadas (los nodos actuales y anteriores)
        log = ""
        for Estacion in self.EstacionesVisitadas:
            for CaminosRestantes in Estacion.CaminosRestantesPorRecorrer:
                log += f"{CaminosRestantes.GetEstaciones()[0]} y {CaminosRestantes.GetEstaciones()[1]}\n"
        print(log)

    #Esta funcion obtiene el camino de menor costo de todos los caminos restantes en toda la lista de
    #EstacionesVisitadas,
    def ObtenerCaminoDeMenorCosto(self):

        PosibleOtraEstacion : CostoEstacionSO = None
        caminoMenosCostoso: CostoCaminoSO = None
        NuevaEstacion: CostoEstacionSO = None
        caminoMejor: CaminoRecorrido = None

        #para todos los caminos restantes
        log :str = ""
        for EstacionVisitada in self.EstacionesVisitadas:
            for Camino in EstacionVisitada.CaminosRestantesPorRecorrer:
                #revisar cual es el camino contrario al que estamos revisando
                if (Camino.EstacionesSO[0] == EstacionVisitada.EstacionInicialSO):

                    PosibleOtraEstacion = Camino.EstacionesSO[1]

                else:

                    PosibleOtraEstacion = Camino.EstacionesSO[0]

                log += f"Posible otra estacion: {PosibleOtraEstacion} en camino: {EstacionVisitada.EstacionInicialSO}\n\t"
                #comparar el costo entre todos los nuevos caminos
                # para cuando es el primer ciclo
                if caminoMenosCostoso == None:

                    caminoMenosCostoso = Camino
                    NuevaEstacion = PosibleOtraEstacion
                    caminoMejor = EstacionVisitada
                    log += f"Primera eleccion de \033[95m{NuevaEstacion}\033[0m\n"

                #Calcular el costo de este nuevo nodo
                log += f"Hacia {PosibleOtraEstacion.Estacion}:Costo acumulado: {EstacionVisitada.CostoCaminoRecorrido}, Costo camino: {Camino.costo}, Costo nodo: {PosibleOtraEstacion.costo},\t total: \033[93m{EstacionVisitada.CostoCaminoRecorrido + Camino.costo + PosibleOtraEstacion.costo}\033[0m\n"

                costoActual:int = caminoMejor.CostoCaminoRecorrido + caminoMenosCostoso.costo + NuevaEstacion.costo
                costoOtroCamino:int = EstacionVisitada.CostoCaminoRecorrido + Camino.costo + PosibleOtraEstacion.costo

                #Compara el mejor camino con este posible nuevo camino, lo asigna si es mejor
                if costoOtroCamino < costoActual:
                    log += f"Cambio de camino de \033[95m{NuevaEstacion}\033[0m a \033[95m{PosibleOtraEstacion}\033[0m\n"
                    caminoMenosCostoso = Camino
                    NuevaEstacion = PosibleOtraEstacion
                    caminoMejor = EstacionVisitada

        #Imprime los datos del mejor camino, y devuelve variables para moverse a otro camino
        print(log)

        log2:str = ""

        log2 += f"\033[94m Camino a seguir: {caminoMenosCostoso.EstacionesSO[0].Estacion} a {caminoMenosCostoso.EstacionesSO[1].Estacion}\033[0m\n";
        log2 += f"\tMenor costo: {caminoMenosCostoso.costo + NuevaEstacion.costo}";
        log2 += f"\tcamino total menor: {caminoMejor.CostoCaminoRecorrido} + {caminoMenosCostoso.costo} + {NuevaEstacion.costo} = {caminoMejor.CostoCaminoRecorrido + caminoMenosCostoso.costo + NuevaEstacion.costo}: Estacion a moverse: \033[96m{NuevaEstacion.Estacion}\033[0m\n";

        print(log2)

        return PosibleOtraEstacion, caminoMenosCostoso, NuevaEstacion, caminoMejor

    def MoverseAOtroNodo(self, caminoMenosCostoso:CostoCaminoSO, NuevaEstacion: CostoEstacionSO, caminoMejor: CaminoRecorrido):
        #agregar caminoMenosCostoso a lista de exclusion
        self.caminosPreviosYaRevisados.append(caminoMenosCostoso)

        #agregar caminoMenosCostoso a CaminosPreviosYaRevisados
        self.caminoMejor.CaminosYaRecorridos.append(caminoMenosCostoso)
        #eliminar camino de CaminosRestantesPorRecorrer[caminoMenosCostoso]
        self.caminoMejor.CaminosRestantesPorRecorrer.remove(caminoMenosCostoso)

        #Cambiar estacionARevisarEnElSiguienteCiclo, agregas el costo total como la suma de el camino mejor y el menos costoso
        self.EstacionARevisarEnElSiguienteCiclo = CaminoRecorrido(NuevaEstacion, [], caminoMejor.ListaEstacionesRecorridasEnEsteCiclo, caminoMenosCostoso.costo + caminoMejor.CostoCaminoRecorrido)
```
